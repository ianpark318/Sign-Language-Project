```python
from glob import glob
from tqdm import tqdm
import json
import os
import torch
import numpy as np
```

```python
morp_path = '/home/jaehyeong/sl/data/morpheme/01'
```

```python
morpheme_list = glob(morp_path + '/*')
morpheme_list.sort()
morpheme_list
```

```python
labels = []
index = 0
for morpheme in tqdm(morpheme_list, desc='Morpheme loading'):
    if index >= 1000:
        break
    with open(morpheme) as json_file:
        data = json.load(json_file)
    l = data['data'][0]['attributes'][0]['name']
    labels.append(l)
    index += 1
```

```python
label_dict = {'labels' : labels}
with open('labels.json', 'w', encoding='UTF-8') as f:
    json.dump(label_dict, f, ensure_ascii=False)
```

### 아래는 데이터 전처리 코드.
1001 번부터 3000번까지 지우고 F만 남기고 U, D, L, R 다 지움.

```python
mp = '/home/jaehyeong/sl/data/morpheme'
kp = '/home/jaehyeong/sl/data/keypoint'
```

```python
mp_list = glob(mp + '/*/*')
for m in mp_list:
    if '_F_morpheme' in m:
        continue
    os.system(f'rm {m}')
    print(f"Removed {m} successfully")
```

```python
mp_list = glob(mp + '/*/*')
for m in mp_list:
    fn = os.path.split(m)[-1]
    if int(fn[11:15]) > 1000:
        os.system(f'rm {m}')
        print(f'Removed {m} successfully')
```

```python
mp_list = glob(mp + '/*')
mp_list.sort()
for user in mp_list:
    print(user, len(os.listdir(user)))
```

```python
kp_list = glob(kp + '/*/*')
for k in kp_list:
    if '_F' in k:
        continue
    os.system(f'rm -rf {k}')
    print(f"Removed {k} successfully")
```

```python
kp_list = glob(kp + '/*/*')
for k in kp_list:
    dn = os.path.split(k)[-1]
    if int(dn[11:15]) > 1000:
        os.system(f'rm -rf {k}')
        print(f'Removed {k} successfully')
```
